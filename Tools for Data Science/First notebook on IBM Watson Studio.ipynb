### The next code prints a pyramid of 5 levels

```python
def get_digits_number(number) :
    count=0
    while(number>0):
        count=count+1
        number=number//10
    return count
        
number = 123456789        
print('The number', str(number), 'has', str(get_digits_number(number)), 'digits')
```
